```python
"""

1'den 1000'e kadar olan asal sayıları ekrana yazdıran bir program yazın. Daha sonra bir tane decorator fonksiyon kullanarak bu fonksiyona 1'den 1000'e kadar olan mükemmel sayıları yazdırma özelliği ekleyin.

"""

def mükemmel_sayı(fonk):
    def tambolen():
        print("mükemmel sayılar")
        for sayı in range(1,1001):
            top=0
            i=1
            while i<sayı:
                if sayı%i==0:
                    top+=i
                i+=1
            if top== sayı:
                print(sayı)

        return fonk()
    return tambolen
@mükemmel_sayı       
def asal_bul():
    print("asal sayılar")
    for sayı in range(2,1001):
        i=2
        say =0
        while i<sayı:
            if sayı %i==0:
                say+=1
            i+=1
        if say==0:
            print(sayı)
asal_bul()
```
